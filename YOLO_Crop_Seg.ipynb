```python
# Training model YOLO

from ultralytics import YOLO

# Load model YOLO
model = YOLO('/content/drive/MyDrive/yolo/yolo11m-seg.pt')

# Start training
model.train(
    data='/content/drive/MyDrive/yolo/crop_map.v5i.yolov11/data.yaml',
    epochs=100,
    imgsz=640,
    batch=8,
    project='/content/drive/MyDrive/yolo/crop_map.v5i.yolov11/result',
    name='yolo11m_paddy_seg',
    device=0,
    # select the augmentation you want to use
    flipud=0.5,      # objects do not have top-down orientation
    degrees=45,      # objects can face any direction
    hsv_h=0.0,       # maintain the color of vegetation

    # Stability training
    optimizer="AdamW",
    lr0=0.001,
    patience=20,
)
```
